# ***Matplotlib***



```python
import matplotlib.pyplot as plt
import numpy as np

xpoint = np.array([0,2,4,6])
ypoint = np.array([0,75,100,250])
plt.title("Title of chart")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.plot(xpoint,ypoint)
plt.show()
```

```python
x = np.array([1,3,5,7])
y = np.array([25,85,140,240])
plt.plot(x,y,marker = "o",ms=15,mec='b',mfc='purple',linestyle="dotted",color = 'black',linewidth=3)
plt.show()
```

# **Scatter plot**

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,104,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show()
```

### **Bar chart**

```python
x = np.array(["A","B","C","D"])
y = np.array([30,58,45,60])
plt.bar(x,y)
plt.show()
```

### **Pie chart**

```python
x = np.array([15,35,20,30])
plt.pie(x)
plt.show()
```
